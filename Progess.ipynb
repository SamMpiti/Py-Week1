```python
from datetime import date
import pandas as pd
import numpy as np
from numpy import genfromtxt
```

```python
data = tv.get_hist(symbol='EURUSD',exchange='OANDA',interval=Interval.in_15_minute, n_bars=10000)
```

```python
df = data.reset_index(drop=False)
```

```python
df['Date'] = df['datetime'].dt.date
df['Time'] = df['datetime'].dt.time
```

```python
start_date = pd.to_datetime('2023-10-31').date()
end_date = pd.to_datetime('2023-11-31').date()
start_time = pd.to_datetime('00:00:00').time()
end_time = pd.to_datetime('23:00:00').time()
```

```python
filtered_df = df[
(df['datetime'].dt.date >= start_date) &
(df['datetime'].dt.date <= end_date) &
(df['datetime'].dt.time >= start_time) &
(df['datetime'].dt.time <= end_time)
]
```

```python
#Bearish orderblock
#Previous candlestick low == next candlestick close
k=1;
for i, row in filtered_df.iloc[1:].iterrows():
    if k!=1:
      if filtered_df.low[k]==filtered_df.close[i]:
        print(f'condition met on \n{filtered_df.datetime[k]} and {filtered_df.datetime[i]}')
        print(f'low=close={filtered_df.low[k]} on {filtered_df.datetime[k]}\n')
    k=i
```

```python
#Bullish orderblock
#Previous candlestick high == next candlestick close
k=1;
for i, row in filtered_df.iloc[1:].iterrows():
    if k!=1:
      if filtered_df.high[k]==filtered_df.close[i]:
        print(f'condition met on \n{filtered_df.datetime[k]} and {filtered_df.datetime[i]}')
        print(f'low=close={filtered_df.close[k]} on {filtered_df.datetime[k]}\n')
    k=i
```
